>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020?
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?

>5) How many entities' largest city is their capital city?
>6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#1
facts['Population_density'] = facts['Pop_2020']/facts['Area_land']
facts1 = facts.sort_values(by ='Population_density', ascending=False)
facts1.head(5)
```

```python
#2
facts[facts['Area_water'] > facts['Area_land']]
```

```python
#3
facts['increase'] = facts['Pop_2020'] - facts['Pop_2010']
facts
facts.sort_values(by='increase',ascending=False).head(5)
```

```python
#4
facts.sort_values(by='Pop_2020',ascending=False).head(5)
#State Bird representing highest Pop as of 2020 is the California quail
```

```python
#4
facts.sort_values(by='Area_land',ascending=False).head(5)
#State Bird representing highest Area Land is the Willow ptarmigan
```

```python
#5
facts.columns
facts[facts['Capital_pop'] > facts['city_1_pop']]
```

```python
#6 Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)
facts['Percent_Drop'] = 100*((facts['city_1_pop'] - facts['city_5_pop'])/facts['city_1_pop'])
facts.sort_values(by='Percent_Drop',ascending=False).head(5)
```

```python
df = pd.read_csv('../data/state_dates.tsv',delimiter="\t")
merged = pd.merge(facts,df ,left_on='USPS_code',right_on='Abbreviation',how='outer')
```

```python
#8
merged['Status'].value_counts()
```

```python
#8
most_dense = merged[merged['Status'] == 'Territory']
territory = most_dense.sort_values(by = 'Population_density', ascending =False)
territory.head(1)
```

```python
#8
least_dense = merged[merged['Status'] == 'State']
state = least_dense.sort_values(by = 'Population_density', ascending =True)
state.head(1)
```
